```python
river = pd.read_csv('data/annotated_points_band4_np_50.csv') 
non_river=pd.read_csv('data/annotated_points_band4_np_150.csv')
river_cor=river.values
non_river_cor=non_river.values
```

```python
intensity1=img1[river_cor[:,0],river_cor[:,1]]       #collecting intensity from each img for river
intensity2=img2[river_cor[:,0],river_cor[:,1]]
intensity3=img3[river_cor[:,0],river_cor[:,1]]
intensity4=img4[river_cor[:,0],river_cor[:,1]]

non_intensity1=img1[non_river_cor[:,0],non_river_cor[:,1]]       #collecting intensity from each img for non-river
non_intensity2=img2[non_river_cor[:,0],non_river_cor[:,1]]  
non_intensity3=img3[non_river_cor[:,0],non_river_cor[:,1]]  
non_intensity4=img4[non_river_cor[:,0],non_river_cor[:,1]]  
```

```python
river_data = {
    'b1': intensity1,
    'b2': intensity2,
    'b3': intensity3,
    'b4': intensity4,
    'label':0
}
df_river = pd.DataFrame(river_data)

non_river_data = {
    'b1': non_intensity1,
    'b2': non_intensity2,
    'b3': non_intensity3,
    'b4': non_intensity4,
    'label':1
}
df_non_river = pd.DataFrame(non_river_data)
```

```python
df_river.to_csv('river_data.csv', index=False)
df_non_river.to_csv('non_river_data.csv', index=False)
```

```python
b1_rmean=df_river['b1'].mean()
b2_rmean=df_river['b2'].mean()
b3_rmean=df_river['b3'].mean()
b4_rmean=df_river['b4'].mean()

b1_nrmean=df_non_river['b1'].mean()
b2_nrmean=df_non_river['b2'].mean()
b3_nrmean=df_non_river['b3'].mean()
b4_nrmean=df_non_river['b4'].mean()

river_mean=np.array([b1_rmean,b2_rmean,b4_rmean,b4_rmean])
non_river_mean=np.array([b1_nrmean,b2_nrmean,b4_nrmean,b4_nrmean])
```

```python
selected_columns = ['b1', 'b2', 'b3', 'b4']
selected_data1 = df_river[selected_columns]
selected_data2 = df_river[selected_columns]

river_cov = np.cov(selected_data1.to_numpy(), rowvar=False)
non_river_cov = np.cov(selected_data1.to_numpy(), rowvar=False)
```

```python
def multivariate_normal(x, mean, cov):
    x = np.array(x)
    mean = np.array(mean)
    cov = np.array(cov)
    
    n = len(mean)
    exponent = -0.5 * np.dot(np.dot((x - mean), np.linalg.inv(cov)), (x - mean).T)
    
    normalization = 1 / ((2 * np.pi) ** (n / 2) * np.linalg.det(cov) ** 0.5)
    
    px_val = normalization * np.exp(exponent)
    
    return px_val
```

```python
def dist_x(x):
    x1=multivariate_normal(x,river_mean,river_cov)
    x2=multivariate_normal(x,non_river_mean,non_river_cov)
    return x1,x2
```

```python
img_combined=np.stack((img1,img2,img3,img4),axis=2)
prior_river=river_cor.shape[0]/(river_cor.shape[0]+non_river_cor.shape[0])
prior_non_river=non_river_cor.shape[0]/(river_cor.shape[0]+non_river_cor.shape[0])
```

```python
final_image = np.zeros((512, 512), dtype=np.uint8)
```

```python

for i in range(512):
    for j in range(512):
        a,b=dist_x(img_combined[i,j,:])
        if prior_river*a >=prior_non_river*b:
            final_image[i][j]='0'
        else:
            final_image[i][j]='1'
```

```python
plt.imshow(final_image,cmap='gray')
plt.imsave('final_image.jpg',final_image,cmap='gray')
```

```python
def confusion_matrix_for_label(true_labels, predicted_labels, label):
    conf_matrix = np.zeros((2, 2), dtype=int)
    for true, pred in zip(true_labels, predicted_labels):
        if true == label and pred == label:
            conf_matrix[0, 0] += 1  # True positive
        elif true == label and pred != label:
            conf_matrix[0, 1] += 1  # False negative
        elif true != label and pred == label:
            conf_matrix[1, 0] += 1  # False positive
        else:
            conf_matrix[1, 1] += 1  # True negative
    return conf_matrix
```

```python
label_river=final_image[river_cor[:,0],river_cor[:,1]]
label_non_river=final_image[non_river_cor[:,0],non_river_cor[:,1]]

output=np.concatenate((label_river.flatten(),label_non_river.flatten()))
input=np.concatenate((df_river['label'],df_non_river['label']))
```

```python
conf_matrix = confusion_matrix_for_label(input,output, 0)
```

```python

view_confusion_matrix = pd.DataFrame(conf_matrix, columns=['Actual River', 'Actual Non-River'], index=['Predicted River', 'Predicted Non-River'])
print(view_confusion_matrix)
```
